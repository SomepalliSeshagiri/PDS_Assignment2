```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
del df[df.columns[0]]
```

```python
df.head()
```

**a) Handling missing values**

```python
df.isnull().sum()
```

**Removing units**

```python
df['Mileage'] = pd.to_numeric(df['Mileage'].str.replace(' km/kg | kmpl','',regex = True),errors='coerce')
df['Engine'] = pd.to_numeric(df['Engine'].str.replace(' CC','',regex = True),errors='coerce')
df['Power'] = pd.to_numeric(df['Power'].str.replace(' bhp','',regex = True),errors='coerce')
df['New_Price'] = pd.to_numeric(df['New_Price'].str.replace(' Lakh | Cr','',regex = True),errors='coerce')
```

```python
df.head()
```

**a) Continuation**

```python
df['Mileage'].fillna(df['Mileage'].mean(),inplace = True)
df['Engine'].fillna(df['Engine'].mean(),inplace = True)
df['Power'].fillna(df['Power'].mean(),inplace = True)
df['Seats'].fillna(df['Seats'].mode()[0],inplace = True)
df['New_Price'].fillna(df['New_Price'].median(),inplace = True)
```

Justification :
Handling missing values means deciding what to do with the empty spots in your data. I filled these gaps with average values from other data points for numerical data and mode for categorical data

```python
df.isnull().sum()
```

**C) One Hot encoding**

```python
df = pd.get_dummies(df,columns=['Fuel_Type','Transmission'] ,prefix=['Fuel_Type','Transmission'])
```

```python
df.head()
```

**d) Creating a New feature**

```python
year_count = datetime.now().year
df['year_count'] = year_count-df['Year']
```

```python
df.head()
```

```python
selected_columns = df[['Mileage', 'Engine', 'Price']]
print("\nSelected Columns:")
print(selected_columns)
```

```python
filtered_data = df[df['Owner_Type'] == 'First']
print("\nFiltered Data:")
print(filtered_data)
```

```python
df.rename(columns={'Make and model': 'Car Model', 'Year of manufacture': 'Manufacture Year'}, inplace=True)
print("\nRenamed DataFrame:")
print(df)
```

```python
df['miles driven'] = df['Kilometers_Driven'] * 0.621371  # Convert kilometers to miles
print("\nMutated DataFrame:")
print(df)
```

```python
sorted_data = df.sort_values(by='Price', ascending=False)
print("\nSorted DataFrame by Price (Descending):")
print(sorted_data)
```

```python
grouped_data = df.groupby('Fuel_Type_Petrol').agg({'Mileage': 'mean', 'Price': ['min', 'max', 'mean']})
print("\nSummary Statistics Grouped by petrol Fuel Type:")
print(grouped_data)
```
